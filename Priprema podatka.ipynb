# Klinički podaci

Sklopićemo klinički set podataka od 2 početna seta: **data_clinical_patient.tsv** i **data_clinical_sample.tsv**.

## data_clinical_patient.tsv

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_patient = pd.read_csv('brca_metabric/data_clinical_patient.tsv', sep='\t')
data_patient.head(7)
```

Značenje kolona:
1. **PATIENT_ID** - id pacijenta
2. **LYMPH_NODES_EXAMINED_POSITIVE** - broj zahvaćenih limfnih čvorova (prati se jer su čvorovi ispod pazuha najčešća mesta prve metastaze)
3. **NPI** - Prognostički indeks koji se računa nakon operacije (zavisi od 3 patološka kriterijuma: veličine tumora, broja zahvaćenih limfnih čvorova i stepena tumora)
4. **CELLULARITY** - celularnost (količina tumorskih ćelija u tkivu pacijenta)
5. **CHEMOPTERAPY** - da li je pacijent podvrgnut hemoterapiji ili ne
6. **COHORT** - broj ispitaničke grupe kojoj pacijent pripada
7. **ER_STATUS** - status biomarkera (Estrogen Receptor: ako postoji velika količina hormonskog receptora u tkivu, onda je pacijent pozitivan za taj biomarker); meri se putem tehinke ImunoHistoChemistry
8. **HER2_SNP6** - status biomarkera HER2 (epidermalni faktor rasta) meren putem tehnike/alata SNP6
9. **HORMONE_TERAPY** - da li je pacijent podvrgnut hormonskoj terapiji ili ne
10. **INFERRED_MENOPAUSAL_STATE** - da li je pacijent u dobu pre ili posle menopauze
11. **INT_CLUST** - integrativni klaster (klasifikator podtippova zasnovan na CNA analizi)
12. **AGE_AT_DIAGNOSIS** - starosna dob u kojoj je pacijentu dijagnostikovan kancer
13. **OS_MONTHS** - koliko je meseci pacijent živeo nakon što mu je dijagnostikovan kancer
14. **OS_STATUS** - da li je osoba još uvek živa ili je preminula (napomena: ako je osoba označena kao preminula, ne mora da znači da je preminula od bolseti - ovaj podatak je opšte prirode)
15. **CLAUDIN_SUBTYPE** - molekularni tip kancera; **CILJNI ATRIBUT**
16. **THREEGENE** - 3-Gene podtip
17. **VITAL_STATUS** - oznaka da li je pacijent živ, a ako ne da li je preminuo od posledica bolesti ili nekog drugog uzroka
18. **LATERALITY** - oznaka da li je obolela leva ili desna dojka
19. **RADIO_THERAPY** - oznaka da li je pacijent podvrgnut radio terapiji ili ne
20. **HISTOLOGYCAL_SUBTYPE** - histološki podtip kancera
21. **BREAST_SURGERY** - tip zahvata nad pacijentom

## Nedostajuće vrednosti kod ostalih atributa

Nedostajuće vrednosti **LYMPH_NODES_EXAMINED_POSITIVE** i **NPI** atributa ćemo zameniti nekim random brojem iz skupa vrednosti. Razlog za to je što sa logičke strane ne dobijamo ništa zamenom prosečnom ili najčešćom vrednošću.

```python
def random_imputation(data, feature):
    number_of_missing_values = data[feature].isna().sum()
    observed_values = data.loc[data[feature].notna(), feature]
    
    new_values = np.random.choice(observed_values, number_of_missing_values)
    
    data.loc[data[feature].isna(), feature] = new_values
    
    return data
```

```python
# Crtanje grafika
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(age_array, wedgeprops=dict(width=0.5), startangle=-40)

ages = [
    "Starost u intevalu: [20, 30]", 
    "Starost u intevalu: [30, 40]", 
    "Starost u intevalu: [40, 50]",
    "Starost u intevalu: [50, 60]",
    "Starost u intevalu: [60, 70]",
    "Starost u intevalu: [70, 80]",
    "Starost u intevalu: [80, 90]",
    "Starost u intevalu: [90, 100]"
]

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(ages[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Age intervals")

plt.show()
```
